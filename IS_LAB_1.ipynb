# **3 JUGS PROBLEM**

```python
from collections import deque

def bfs_jug_solver(capacities, target_amount):
    def generate_next_states(state):
        a, b, c = state
        cap_a, cap_b, cap_c = capacities

        return [
            # Fill each jug
            (cap_a, b, c),
            (a, cap_b, c),
            (a, b, cap_c),

            # Empty each jug
            (0, b, c),
            (a, 0, c),
            (a, b, 0),

            # Pour between jugs
            (a - min(a, cap_b - b), b + min(a, cap_b - b), c),
            (a - min(a, cap_c - c), b, c + min(a, cap_c - c)),
            (a + min(b, cap_a - a), b - min(b, cap_a - a), c),
            (a, b - min(b, cap_c - c), c + min(b, cap_c - c)),
            (a + min(c, cap_a - a), b, c - min(c, cap_a - a)),
            (a, b + min(c, cap_b - b), c - min(c, cap_b - b))
        ]

    initial_state = (capacities[0], 0, 0)  # Start with the first jug full
    queue = deque([(initial_state, [])])
    visited = set([initial_state])

    while queue:
        state, path = queue.popleft()

        if target_amount in state:
            return path + [state]

        for next_state in generate_next_states(state):
            if next_state not in visited:
                visited.add(next_state)
                queue.append((next_state, path + [state]))

    return None  # No solution found

def main():
    try:
        # Get user input for jug capacities and target amount
        cap_a = int(input("Enter the capacity of the first jug: "))
        cap_b = int(input("Enter the capacity of the second jug: "))
        cap_c = int(input("Enter the capacity of the third jug: "))
        target_amount = int(input("Enter the target amount of water: "))

        capacities = (cap_a, cap_b, cap_c)

        # Check if the target amount is achievable with the given capacities
        if target_amount > max(capacities):
            print("The target amount is greater than the capacity of the largest jug.")
            return

        solution = bfs_jug_solver(capacities, target_amount)

        if solution:
            print("Solution found:")
            for state in solution:
                print(state)
        else:
            print("No solution exists.")

    except ValueError:
        print("Please enter valid integer values for capacities and target amount.")

if __name__ == "__main__":
    main()
```

# **8 PUZZLE PROBLEM**

```python
from heapq import heappop, heappush

class PuzzleState:
    def __init__(self, board, moves=0, prev=None):
        self.board = board
        self.blank_pos = board.index(0)
        self.moves = moves
        self.prev = prev

    def __lt__(self, other):
        return (self.moves + self.manhattan_distance()) < (other.moves + other.manhattan_distance())

    def manhattan_distance(self):
        return sum(
            abs(x - goal_x) + abs(y - goal_y)
            for i, tile in enumerate(self.board)
            if tile != 0
            for goal_x, goal_y in [divmod(tile - 1, 3)]
            for x, y in [divmod(i, 3)]
        )

    def get_neighbors(self):
        x, y = divmod(self.blank_pos, 3)
        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
        neighbors = []

        for dx, dy in directions:
            new_x, new_y = x + dx, y + dy
            if 0 <= new_x < 3 and 0 <= new_y < 3:
                new_blank_pos = new_x * 3 + new_y
                new_board = list(self.board)
                new_board[self.blank_pos], new_board[new_blank_pos] = new_board[new_blank_pos], new_board[self.blank_pos]
                neighbors.append(PuzzleState(tuple(new_board), self.moves + 1, self))

        return neighbors

    def is_goal(self):
        return self.board == (1, 2, 3, 8, 0, 4, 7, 6, 5)

def a_star_search(start):
    open_set, visited = [start], set()

    while open_set:
        current = heappop(open_set)

        if current.is_goal():
            return current

        visited.add(current.board)

        for neighbor in current.get_neighbors():
            if neighbor.board not in visited:
                heappush(open_set, neighbor)

    return None

def print_state(board):
    return f"{board[:3]}\n{board[3:6]}\n{board[6:]}"

def print_solution(solution):
    path = []
    while solution:
        path.append(solution.board)
        solution = solution.prev

    for i, step in enumerate(reversed(path), 1):
        print(f"STEP {i}:")
        print(print_state(step))
        print()

# Example usage
start_board = (2, 8, 3, 1, 6, 4, 0, 7, 5)  # Start state
goal_board = (1, 2, 3, 8, 0, 4, 7, 6, 5)   # Goal state

print("Start State:")
print(print_state(start_board))
print("\nGoal State:")
print(print_state(goal_board))

solution = a_star_search(PuzzleState(start_board))

if solution:
    print("\nSolution found:\n")
    print_solution(solution)
else:
    print("No solution exists.")
```

# **TRAVELLING SALESMAN PROBLEm**

```python
import random
import math

def convert_distance_matrix(distance_matrix):
    """Convert string-based distance matrix to numeric values."""
    # Create a mapping from string to numerical values for distances
    distance_map = {
        'A': 1,
        'B': 2,
        'C': 3,
        'D': 4,
        'E': 5
    }

    numeric_matrix = [[distance_map[val] for val in row] for row in distance_matrix]
    return numeric_matrix

def calculate_total_distance(tour, distance_matrix):
    """Calculate the total distance of the given tour."""
    return sum(distance_matrix[tour[i]][tour[i + 1]] for i in range(len(tour) - 1)) + distance_matrix[tour[-1]][tour[0]]

def simulated_annealing(distance_matrix, initial_temperature=1000, cooling_rate=0.995, num_iterations=10000):
    n = len(distance_matrix)

    # Generate an initial random tour
    current_tour = list(range(n))
    random.shuffle(current_tour)
    current_distance = calculate_total_distance(current_tour, distance_matrix)

    best_tour = current_tour
    best_distance = current_distance

    temperature = initial_temperature

    for _ in range(num_iterations):
        # Generate a new tour by swapping two cities
        new_tour = current_tour[:]
        i, j = random.sample(range(n), 2)
        new_tour[i], new_tour[j] = new_tour[j], new_tour[i]
        new_distance = calculate_total_distance(new_tour, distance_matrix)

        # Decide whether to accept the new tour
        if new_distance < current_distance or random.random() < math.exp((current_distance - new_distance) / temperature):
            current_tour = new_tour
            current_distance = new_distance

            if new_distance < best_distance:
                best_tour = new_tour
                best_distance = new_distance

        # Cool down
        temperature *= cooling_rate

    return best_tour, best_distance

def print_matrix(matrix):
    """Print the matrix in a readable format."""
    for row in matrix:
        print(" ".join(map(str, row)))
    print()

# Example distance matrix with string values
string_distance_matrix = [
    ['A', 'B', 'C', 'D', 'E'],
    ['A', 'B', 'C', 'D', 'E'],
    ['A', 'B', 'C', 'D', 'E'],
    ['A', 'B', 'C', 'D', 'E'],
    ['A', 'B', 'C', 'D', 'E']
]

# Convert string-based matrix to numeric values
numeric_distance_matrix = convert_distance_matrix(string_distance_matrix)

print("Distance Matrix:")
print_matrix(string_distance_matrix)

print("Distance Matrix:")
print_matrix(numeric_distance_matrix)

# Perform simulated annealing to find the best tour
best_tour, best_distance = simulated_annealing(numeric_distance_matrix)
print("Best tour:", best_tour)
print("Best distance:", best_distance)
```
